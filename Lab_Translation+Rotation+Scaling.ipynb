```python
import math
import matplotlib.pyplot as plt

def rotation(x, y, angle, xp, yp):
    '''Apply 2D rotation to polygon vertices'''
    radian = math.radians(angle)
    cos_a = math.cos(radian)
    sin_a = math.sin(radian)
    
    x_new, y_new = [], []
    for i in range(len(x)):
        # Translate to origin
        x_shift = x[i] - xp
        y_shift = y[i] - yp
        
        # Rotate
        x_rot = x_shift * cos_a - y_shift * sin_a
        y_rot = x_shift * sin_a + y_shift * cos_a
        
        # Translate back
        x_new.append(xp + x_rot)
        y_new.append(yp + y_rot)
    
    return x_new, y_new


def scaling(x, y, sx, sy, xp, yp):
    '''Apply 2D scaling to polygon vertices'''
    x_new, y_new = [], []
    for i in range(len(x)):
        # Translate to origin
        x_shift = x[i] - xp
        y_shift = y[i] - yp
        
        # Scale
        x_scaled = x_shift * sx
        y_scaled = y_shift * sy
        
        # Translate back
        x_new.append(xp + x_scaled)
        y_new.append(yp + y_scaled)
    
    return x_new, y_new


def translation(x, y, tx, ty):
    '''Apply 2D translation to polygon vertices'''
    x_new = [x[i] + tx for i in range(len(x))]
    y_new = [y[i] + ty for i in range(len(y))]
    return x_new, y_new


def main():
    # Triangle vertices (original)
    x = [100, 200, 150]
    y = [100, 100, 200]
    
    # Transformation parameters
    angle = 45          # degrees
    sx, sy = 1.5, 1.5   # scale factors
    tx, ty = 80, 80     # translation distances
    xp, yp = 150, 150   # pivot point
    
    # Apply transformations
    x_rot, y_rot = rotation(x, y, angle, xp, yp)
    x_scale, y_scale = scaling(x, y, sx, sy, xp, yp)
    x_trans, y_trans = translation(x, y, tx, ty)
    
    # Plot
    plt.figure(figsize=(10, 5))
    
    # Original triangle
    plt.plot(x + [x[0]], y + [y[0]], 'k-',linewidth=3, label='Original')
    
    # Rotated triangle
    plt.plot(x_rot + [x_rot[0]], y_rot + [y_rot[0]], 'r--',linewidth=3, label=f'Rotated {angle}')
    
    # Scaled triangle
    plt.plot(x_scale + [x_scale[0]], y_scale + [y_scale[0]], 'b-.',linewidth=3, label=f'Scaled ({sx}, {sy})')
    
    # Translated triangle
    plt.plot(x_trans + [x_trans[0]], y_trans + [y_trans[0]], 'g:',linewidth=3, label=f'Translated ({tx}, {ty})')
    
    # Pivot point
    plt.plot(xp, yp, 'go', markersize=8, label='Pivot Point')
    
    # Formatting
    plt.axis('equal')
    plt.grid(True)
    plt.legend()
    plt.title('2D Transformations: Rotation, Scaling, and Translation')
    plt.xlabel('X-axis')
    plt.ylabel('Y-axis')
    plt.show()


if __name__ == "__main__":
    main()
```
