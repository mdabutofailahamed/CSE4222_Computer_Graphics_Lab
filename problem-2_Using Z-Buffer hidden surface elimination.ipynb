```python
import pygame
import sys
import numpy as np

# Initialize Pygame
pygame.init()

# Colors (R, G, B)
YELLOW = (255, 255, 0)
RED = (255, 0, 0)
GREEN = (0, 255, 0)
BLACK = (0, 0, 0)

# Set up window
win_width = 400
win_height = 400
screen = pygame.display.set_mode((win_width, win_height))
pygame.display.set_caption("Shape Drawer with Z-buffer")

# Fill background with black
screen.fill(BLACK)

# --- Z-buffer initialization ---
zbuffer = np.full((win_width, win_height), -float('inf'))  # সব depth = -inf (background)

# --- Helper function to plot pixel with depth check ---
def plot_pixel(x, y, color, z):
    if 0 <= x < win_width and 0 <= y < win_height:
        if z > zbuffer[x, y]:  # Z-buffer check
            screen.set_at((x, y), color)
            zbuffer[x, y] = z

# --- Shape functions with depth ---
def draw_triangle(z=0):
    x = [10, 50, 100]
    y = [100, 20, 100]
    points = [(x[i], y[i]) for i in range(3)]
    # Simple scanline fill
    min_y = min(y)
    max_y = max(y)
    for py in range(min_y, max_y+1):
        # Find intersections with triangle edges
        intersections = []
        for i in range(3):
            x0, y0 = points[i]
            x1, y1 = points[(i+1)%3]
            if y0 == y1:
                continue
            if y0 <= py <= y1 or y1 <= py <= y0:
                x_int = int(x0 + (py - y0) * (x1 - x0) / (y1 - y0))
                intersections.append(x_int)
        intersections.sort()
        for i in range(0, len(intersections), 2):
            for px in range(intersections[i], intersections[i+1]+1):
                plot_pixel(px, py, YELLOW, z)

def draw_circle(z=0):
    cx, cy, r = 100, 100, 45
    for x in range(cx - r, cx + r + 1):
        for y in range(cy - r, cy + r + 1):
            if (x - cx)**2 + (y - cy)**2 <= r**2:
                plot_pixel(x, y, RED, z)

def draw_rectangle(z=0):
    rect_x1, rect_y1 = 80, 80
    rect_x2, rect_y2 = 180, 180
    for x in range(rect_x1, rect_x2):
        for y in range(rect_y1, rect_y2):
            plot_pixel(x, y, GREEN, z)

# --- Main ---
sequence = input("Enter shape sequence (C, R, T): ")

# Assign depth to each shape based on input order
for i, char in enumerate(sequence):
    z = i  # Later shapes have higher z → appear in front
    if char == 'C':
        draw_circle(z)
    elif char == 'R':
        draw_rectangle(z)
    else:
        draw_triangle(z)

# Update the display
pygame.display.update()

# Wait until user closes the window
while True:
    for event in pygame.event.get():
        if event.type == pygame.QUIT:
            pygame.quit()
            sys.exit()
```
